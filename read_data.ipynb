## SQL Connection

```python
#!pip install sqlalchemy
```

```python
from sqlalchemy import create_engine
```

```python
temp_db = create_engine("sqlite:///:memory:")
```

```python
temp_db.table_names()
```

```python
tables = pd.read_html('https://en.wikipedia.org/wiki/World_population')
```

https://en.wikipedia.org/wiki/World_population

```python
tables
```

```python
tables[5]
```

```python
tables[5].to_sql(name='populations',con=temp_db)
```

```python
temp_db.table_names()
```

```python
df=pd.read_sql(sql="populations",con = temp_db,index_col="index")
```

```python
df
```

```python
pd.read_sql_query(sql = "SELECT Country, Population FROM populations",con = temp_db)
```

```python
df[["Country","Population"]]
```

```python
import sqlite3
```

```python
with sqlite3.connect('sample_data1.db') as cnnt:
    df.to_sql("sample7",cnnt)
```

```python
query1="SELECT Country, Population FROM sample7"
```

```python
with sqlite3.connect("sample_data1.db") as ct:
    df6=pd.read_sql(query1,ct)
```
